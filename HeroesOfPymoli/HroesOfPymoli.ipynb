```python
import pandas as pd
import numpy as np

# File to Load
file_to_load = "c:/Users/Demo/Desktop/Data_Analytics/GitRepo/pandas-challenge/Resources/purchase_data.csv"

Purchdata = pd.read_csv(file_to_load)
```

```python
# Creating series 
purchid=Purchdata["Purchase ID"]
sn=Purchdata["SN"]
age=Purchdata["Age"]
gender=Purchdata["Gender"]
itemid=Purchdata["Item ID"]
itemname=Purchdata["Item Name"]
price=Purchdata["Price"]
```

```python
# getting lenght of series SN
n=len(sn.unique())
```

```python
# Creating Dataframe
player_df=pd.DataFrame({"Total Players":[n]})
player_df
```

```python
#Purchasing Analisys
```

```python
#Aplying methods to get number of items, totals, and average
numitem=len(itemname.unique())
itemcnt=itemid.count()
avgprice=price.mean()
Totalrev=price.sum()
```

```python
#Gender Demographic
```

```python
#Creating new Dataset to organize data and to get Gender data
Result1 = Purchdata.groupby('Gender')
Tcountgdr=Result1.nunique()["SN"]
Playper=Tcountgdr/n
```

```python
#Creating Dataframe
pd.options.display.float_format = '{:,.2f}%'.format
gender_df=pd.DataFrame({'Total Count' : Tcountgdr, 'Percentaje of Players':Playper*100}, index=['Male', 'Female', 'Other / Non-Disclosed'])
gender_df.index.name='Gender'
gender_df
```

```python
#Purchasing Analysis
```

```python
# checking on Result1 values to get Purchasimg data
print(Result1.nunique())
```

```python
#Extracting data from Result1 and perform calculations for averageand totals
AvgPurch=Result1["Purchase ID"].count()
AvgPurchPrice=Result1["Price"].mean()
AvgPurchTot=Result1["Price"].sum()
AvgPurchPP=AvgPurchTot/Tcountgdr
```

```python
#Creating Dataframe
pd.options.display.float_format = '{:,.2f}'.format
gender_df=pd.DataFrame({'Purchase Count' : AvgPurch, 'Average Purchase Price':AvgPurchPrice, 'Total Purchase Value' : AvgPurchTot, 'Avg Total Purchase per person':AvgPurchPP}, index=['Male', 'Female', 'Other / Non-Disclosed'])
gender_df.index.name='Gender'
gender_df
```

```python
# Age Demographics
```

```python
#Creating bins to divide age column
Ages = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 199]
groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
# Creating additional column for AgeByGroup Values, Checking the result
Purchdata["AgeByGroup"]= pd.cut(Purchdata["Age"], Ages, labels=groups, include_lowest=True)
Purchdata
```

```python
# Gettiung the values and making calculations for Age
GroupAge=Purchdata.groupby("AgeByGroup")
TotAgeCnt=GroupAge["SN"].nunique()
AgePercent=TotAgeCnt/n*100
```

```python
#Creating Dataframe
pd.options.display.float_format = '{:,.2f}%'.format
Agedemo_df=pd.DataFrame({'Total Count' : TotAgeCnt, 'Percentage of Players':AgePercent})
Agedemo_df
```

```python
#Purchasing Analysis
```

```python
#Based on the same grouping data, extract info and made calculation for purshasing
PurchCntAge=GroupAge["Purchase ID"].count()
AvgPurchPriceG = GroupAge["Price"].mean()
TotPurchVal = GroupAge["Price"].sum()
AvgPurchPPA = TotPurchVal/ TotAgeCnt
```

```python
#Create Dataframe
pd.options.display.float_format = '${:,.2f}'.format
Purchase_df=pd.DataFrame({'Purchase Count' : PurchCntAge, 'Average Purchase Price':AvgPurchPriceG, 'Total Purchase Value' : TotPurchVal, 'Avg Total Purchase per person':AvgPurchPPA})
Purchase_df.index.name='Age Ranges'
Purchase_df
```

```python
#Top Senders
```

```python
# Now, Getting data from the same set to find best buyers
TSSpend=Purchdata.groupby("SN")
SpendCnt=TSSpend["Purchase ID"].count()
SpendPriceAvg=TSSpend["Price"].mean()
SpendTot=TSSpend["Price"].sum()
```

```python
#Creating dataframe
pd.options.display.float_format = '${:,.2f}'.format
Spend_df=pd.DataFrame({'Purchase Count' : SpendCnt, 'Average Purchase Price':SpendPriceAvg, 'Total Purchase Value' : SpendTot})
Spend_df.index.name='SN'
Spend_df.sort_values(by='Total Purchase Value', ascending=False).head()
```

```python
#Most Popular Items
```

```python
# getting data and calculating for most sold items
PopItem=Purchdata[["Item ID", "Item Name", "Price"]]
ItemIdNam=PopItem.groupby(["Item ID","Item Name"])
ItemPurCnt=ItemIdNam["Price"].count()
ItemPurVal=ItemIdNam["Price"].sum()
ItemPrice=ItemPurVal/ItemPurCnt
```

```python
# Creating dataframe
pd.options.display.float_format = '${:,.2f}'.format
Item_df=pd.DataFrame({'Purchase Count' : ItemPurCnt, 'Item Price':ItemPrice, 'Total Purchase Value' : ItemPurVal})
Item_df.index.name='Item ID'
Item_df.sort_values(by='Purchase Count', ascending=False).head()
```
